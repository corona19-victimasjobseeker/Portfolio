```python
#geopy검색을 방해하는 끝글자를 제거하는 코드 작성
C지역맛집_외지인_주소 = []
for i in 지역맛집_외지인_주소:
    if i[-1] == '0':
        C지역맛집_외지인_주소.append(i[:-2])
    else :
        C지역맛집_외지인_주소.append(i)
C지역맛집_외지인_주소
```

```python
geo_local = Nominatim(user_agent='South Korea')
```

```python
geo = geo_local.geocode('강원 강릉시 초당순두부길77번길 15') #none값이 나오는 이유는 알 수가 없다.
print(geo)
```

```python
geo2 = geo_local.geocode('강원 강릉시 해안로 1459') #전부 다 오류인 것은 아닌듯하다.
print(geo2)
```

```python
#위도, 경도를 반환하는 함수
def geocoding(address):
    geo = geo_local.geocode(address)
    x_y = [geo.latitude, geo.longitude]
    return x_y

geocoding('강원 강릉시 경강로2255번길 21')
```

```python
latitude = []
longtitude = []

for i in C지역맛집_외지인_주소:
    if geo_local.geocode(i) == None:
        latitude.append(0)
        longtitude.append(0)
    else:
        latitude.append(geocoding(i)[0])
        longtitude.append(geocoding(i)[1])
print(latitude, longtitude)
```

### DataFrame을 반복문에 넣으면 <u>**반복되는 값은 컬럼이지. 그 열의 내용이 아니다?**</u>

rg_mat_tour_add의 특정 주소의 컬럼 [lat, lng]에는 0이라는 빈 값이 들어있다.<br>
nan_rg_add에는 위 자료의 빈 값의 주소와 수정된 [lat, lng]가 들어있다. <br>
rg_mat_tour_add의 컬럼[lat, lng]의 빈 값을 지닌 행을 nan_rg_add의 값으로 대체하는 코딩을 작성해보자


```python
rg_mat_tour_add = rg_mat_tour_add.drop(columns='Unnamed: 0')
```

```python
rg_mat_tour_add.head()
```

```python
lat =[37.7911361, 37.6965004, 37.8223696, 37.7713400, 37.8592616, 37.7913873, 37.7028444, 37.7409999, 37.7912484]
lng = [128.914702, 128.891996, 128.884983, 128.948532, 128.831679, 128.914766, 129.015854, 128.983663, 128.915000]
```

```python
lat.index(37.8223696)
```

```python
# DataFrame의 index 속성은 DataFrame의 맨 위 행에서 맨 아래 행까지 범위 객체를 제공합니다.
rg_mat_tour_add.index
```

```python
# rg_mat_tour_add의 컬럼['lat', 'lng']의 값들 중 0인 값을 nan_rg_add의 값들로 바꾸는 코딩
for i in range(len(rg_mat_tour_add)):  # range(len(rg_mat_tour_add)) = range(0, 97) 객체
    if rg_mat_tour_add['lat'][i] == 0: # nan_rg_add.index = Int64Index([0, 2, 11, 27, 46, 50, 61, 71, 81], dtype='int64')
        rg_mat_tour_add['lat'][i] = nan_rg_add['lat'][i]
        rg_mat_tour_add['lng'][i] = nan_rg_add['lng'][i]
    else:
        pass
```
