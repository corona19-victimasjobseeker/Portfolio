```python
gangneung_map3 = folium.Map(
    location = [37.751853, 128.8760574],
    tiles = 'cartodbdark_matter',
    zoom_start = 11)
```

```python
gangneung_map2 = folium.Map(
    location = [37.751853, 128.8760574],
    tiles = 'cartodbdark_matter',
    zoom_start = 11)

for idx in gangneung.index:
    lat = gangneung.loc[idx, '위도']
    lng = gangneung.loc[idx, '경도']
    typ = gangneung.loc[idx, '분류']
    
    fillColor = ''
    if typ == '한식':
        fillColor = 'yellow'
        size = 5
    elif typ == '카페':
        fillColor = 'purple'
        size = 5
    elif typ == '음식점기타':
        fillColor = 'blue'
        size = 5
    elif typ == '제과점':
        fillColor = 'green'
        size = 5
    elif typ == '중식':
        fillColor = 'red'
        size = 5
    elif typ == '패스트푸드':
        fillColor = 'orange'
        
    folium.CircleMarker(
        location= [lat, lng],
        color = fillColor,
        fill = True,
        fill_color = fillColor,
        fill_opacity = 1,
        weight = 1,
        radius = size
    ).add_to(gangneung_map)
    
gangneung_map
```

```python
gangneung_map.save('./gangnueng_cate.html')
```

```python
검색수평균 #13851 gangneung['외지인 검색 수'].mean()
for idx in gangneung.index:
    lat = gangneung.loc[idx, '위도']
    lng = gangneung.loc[idx, '경도']
    typ = gangneung.loc[idx, '분류']
    mean = gangneung.loc[idx, '외지인 검색 수']
    
    if mean > 검색수평균:
        fillColor = 'red'
    else:
        fillColor = 'blue'

    folium.CircleMarker(
        location= [lat, lng],
        color = 'yellow',
        fill = True,
        fill_color = fillColor,
        fill_opacity = 1,
        weight = 1,
        radius = mean/2600
    ).add_to(gangneung_map3)
    
gangneung_map3
```

```python
gangneung_map3.save('./gangnueng_freq.html')
```
